```python
import sympy
from sympy.plotting import plot
```

```python
x = sympy.Symbol ('x')
```

```python
f_x = -5 * x ** 2 + 10 * x - 150
print(f_x)
```

1. Определить корни

```python
for root in sympy.solve(f_x, x):
    print(root.evalf())
```

```python
print([j.evalf() for j in sympy.solve(f_x)])
```

4. Построить график

```python
f_diff_x = f_x.diff()
plot(f_x, f_diff_x)
```

5. Вычислить вершину

```python
plot(f_diff_x, xlim = (0, 2))
```

```python
# for i in range(75, 125):
#     param = i / 100
#     print(param, f_diff_x.subs(x, param))
```

```python
top_x = sympy.solve(f_diff_x)
# print(top_x)
# print(type(top_x))
top_y = f_x.subs(x, top_x[0])
# print(top_y)
print(f"Вершина находится в точке {[top_x[0].evalf(), top_y.evalf()]}")
```
